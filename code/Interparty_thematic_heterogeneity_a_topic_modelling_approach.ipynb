```python
import pandas as pd
import numpy as np
from gensim.models import LdaModel
from gensim.corpora.dictionary import Dictionary
import pickle
import pyLDAvis
import pyLDAvis.gensim_models
```

```python
# Load the preprocessed pre-COVID and during-COVID datasets into two pandas dataframe respectively
# and store the tweet corpora into two pandas series.
path_pre="tweets_pre_covid_preprocessed_topic_modeling.csv"
path_during="tweets_during_covid_preprocessed_topic_modeling.csv"
df_pre=pd.read_csv(path_pre,header=0)
df_during=pd.read_csv(path_during,header=0)
tweets_pre=df_pre["tweet_text"].map(lambda x:eval(x))
tweets_during=df_during["tweet_text"].map(lambda x:eval(x))
```

```python
# Generate the token-id mappings for each document in each of the two corpora using Dictionary
# from gensim. 
dictionary_pre=Dictionary(tweets_pre)
dictionary_during=Dictionary(tweets_during)
```

```python
# Represent each document in each of the two corpora using the BoW format.
training_corpus_pre=[dictionary_pre.doc2bow(doc) for doc in tweets_pre]
training_corpus_during=[dictionary_during.doc2bow(doc) for doc in tweets_during]
```

```python
# Load the optimal LDA models trained on the pre-COVID corpus and the during-COVID corpus.
lda_pre=LdaModel.load("optimal_lda_model_pre_COVID_corpus.model")
lda_during=LdaModel.load("optimal_lda_model_during_COVID_corpus.model")
```

```python
pyLDAvis.enable_notebook()
LDAvis_prepared_pre=pyLDAvis.gensim_models.prepare(lda_pre,training_corpus_pre,dictionary_pre)
```

```python
pyLDAvis.display(LDAvis_prepared_pre)
```

```python
pyLDAvis.enable_notebook()
LDAvis_prepared_during=pyLDAvis.gensim_models.prepare(lda_during,training_corpus_during,dictionary_during)
```

```python
pyLDAvis.display(LDAvis_prepared_during)
```
